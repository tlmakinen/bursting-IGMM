```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.table import Table, Column
```

```python
# Use Anto's functions to estimate and crop out the background around each spot
```

```python
import   InterpolatMissingSpots
import   SpotsBackgroundEstimation
import   AnalysisLoader

foldername = r'C:\Users\tmakinen\repositories\bursting-IGMM\autocorrelation\spot-analysis\AnalysisFolderForAutocorrelation'
filedata         =  AnalysisLoader.RawData(foldername)
spots_3D       =  AnalysisLoader.SpotsIntsVol(foldername)
features_3D  =  AnalysisLoader.Features(foldername)

spots_tracked_3D  =  np.fromfile(foldername + '/spots_tracked.bin', 'uint16')
spots_tracked_3D  =  spots_tracked_3D[3:].reshape(spots_tracked_3D[2], spots_tracked_3D[1], spots_tracked_3D[0])
steps                     =  spots_tracked_3D.shape[0]

```

```python
spt_idxs      =  np.unique(spots_tracked_3D)[1:]           # list of all the spot indexes
```

```python
sp_tag    =  17    # put the id of the spot you want to study
numb_zrs  =  4     # number of consecutive zeros you consider as fake

spot_trace    =  InterpolatMissingSpots.InterpolatMissingSpots(filedata.green4D, spots_tracked_3D, spots_3D, sp_tag, numb_zrs)
bkg_est       =  SpotsBackgroundEstimation.SpotsBackgroundEstimation(spot_trace.spots_tracked_3D, filedata.green4D, sp_tag).bkg_values
```

```python
spt_idxs
```

```python
# run background estimation and cleanup on all spot traces

numb_zrs = 4        # max number of zeros that indicates a "fake" feature
trace_matrix = []   # initialize list of corrected spot traces


for tag in spt_idxs:
    # for each spot, interpolate any missing spots. Then, estimate the background for each corrected trace.    
    spot_trace    =  InterpolatMissingSpots.InterpolatMissingSpots(filedata.green4D, 
                                                                   spots_tracked_3D, spots_3D, tag, numb_zrs) 
    bkg_est       =  SpotsBackgroundEstimation.SpotsBackgroundEstimation(spot_trace.spots_tracked_3D, 
                                                                         filedata.green4D, tag).bkg_values
    
    # Then, add the ratio of spot signal to background spot_trace / bkg to the matrix
    
    div = spot_trace.corrected_trace / bkg_est    # remove nans from the quotient of signal / background
    for i in range(len(div)):
        if np.isnan(div[i]):
            div[i] = 0
    
    trace_matrix.append(div)
    
    #print(tag)
```

```python
tr = np.asarray(trace_matrix)
```

```python
for i in trace_matrix:
    plt.plot(i)
plt.show()
```

```python
# now let's put all of these into a table object

t = Table()

for i in range(len(spt_idxs)):
    label = 'Spot_' + str(spt_idxs[i])
    t[label] = Column(trace_matrix[i], unit='gfp intensity', description='spot intensity')
```

```python
# save table as a csv file
output = t.copy()
output.write('bkg-corrected-spots.csv', overwrite=True)
```
